#### Import necessary libraries

```python
import re
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import FunctionTransformer
```

```python
def clean_name(name_series):
    # Ensure the input is a pandas Series
    if not isinstance(name_series, pd.Series):
        raise TypeError("Expected a pandas Series")

    # Apply cleaning operations
    cleaned_names = name_series.str.replace(r'Writtenby:', '', regex=True)  # Remove 'Writtenby:'
    cleaned_names = cleaned_names.str.replace(r'Narratedby:', '', regex=True)  # Remove 'Writtenby:'
    spaced_names = cleaned_names.str.replace(r'([a-z])([A-Z])', r'\1 \2', regex=True)  # Add spaces

    return spaced_names.to_frame()

clean_name_transformer = FunctionTransformer(clean_name)
```

```python
clean_name_transformer.transform(df['author'])
```

```python
clean_name_transformer.transform(df['narrator'])
```

```python
# Parse hours and minutes for each entry in the Series
def parse_time_string(time):
    hours_match = re.search(r'(\d+)\s*hrs?', time)
    mins_match = re.search(r'(\d+)\s*mins?', time)
    
    # Extract hours and minutes, default to 0 if not found
    hours = int(hours_match.group(1)) if hours_match else 0
    minutes = int(mins_match.group(1)) if mins_match else 0
    
    # Calculate total minutes
    return hours * 60 + minutes

def convert_to_minutes_series(time_series):
    # Ensure the input is a pandas Series
    if not isinstance(time_series, pd.Series):
        raise TypeError("Expected a pandas Series")
    
    return time_series.apply(parse_time_string).to_frame()

convert_to_minutes_transformer = FunctionTransformer(convert_to_minutes_series)
```

```python
from datetime import datetime

def convert_to_ddmmyyyy(date): 
    formatted_date = None
    try:
        # Try parsing with both possible formats
        if '/' in date:
            parsed_date = datetime.strptime(date, "%d/%m/%Y") if len(date.split('/')[-1]) == 4 else datetime.strptime(date, "%d/%m/%y")
        elif '-' in date:
            parsed_date = datetime.strptime(date, "%d-%m-%Y") if len(date.split('-')[-1]) == 4 else datetime.strptime(date, "%d-%m-%y")
        else:
            raise ValueError("Unrecognized date format")
    except ValueError:
        raise TypeError("Expected a pandas Series")
        return None
    else: 
        # Format date to dd/mm/yyyy
        formatted_date = parsed_date.strftime("%d/%m/%Y") 
    return formatted_date

def date_converter(date_series):

    # Ensure the input is a pandas Series
    if not isinstance(date_series, pd.Series):
        raise TypeError("Expected a pandas Series")

    return date_series.apply(convert_to_ddmmyyyy).to_frame()

date_transformer = FunctionTransformer(date_converter)
```

```python
def process_stars_data(stars_string): 
    if "Not rated yet" in stars_string:
        # Handle "Not rated yet"
        stars = ratings = 'Not Rated'
    else:
        # Extract stars (e.g., "4.5" or "5") and ratings (e.g., "34")
        stars_match = re.search(r'(\d+(\.\d+)?) out of 5 stars', stars_string)
        ratings_match = re.search(r'(\d+)\s+ratings', stars_string)

        stars = stars_match.group(1) if stars_match else "Not Rated"
        ratings = ratings_match.group(1) if ratings_match else "Not Rated" 
        
    return stars, ratings


def stars_converter(stars_series):
    # Ensure the input is a pandas Series
    if not isinstance(stars_series, pd.Series):
        raise TypeError("Expected a pandas Series")

    parsed_stars = stars_series.apply(process_stars_data)

    return pd.DataFrame(parsed_stars.tolist(), columns=['rating_stars','ratings'])

stars_transformer = FunctionTransformer(stars_converter)
```

```python
df['stars'].apply(process_stars_data)
```

```python
stars_transformer.transform(df['stars'])
```

Apply Columns Trasnformers to clean all columns data 

```python
# Define the ColumnTransformer
column_transformer = ColumnTransformer(
    transformers=[
        ("clean_author", clean_name_transformer, "author"),
        ("clean_narrator", clean_name_transformer, "narrator"),
        ("total_minutes", convert_to_minutes_transformer, "time"),
        ("formatted_date", date_transformer, "releasedate"),
        ("stars_data", stars_transformer, "stars"),
    ],
    remainder="passthrough"  # Keep any other columns untouched
)
```

```python
tramsformed_columns = column_transformer.fit_transform(df)
transformed_df=pd.DataFrame(tramsformed_columns, columns=[
        "clean_author", "clean_narrator", "total_minutes", 
        "formatted_date", "stars", "ratings", *df.drop(["author", "narrator", "time", "releasedate", "stars"], axis=1).columns
    ])
transformed_df
```

```python
transformed_df['language'] = transformed_df['language'].str.capitalize()
transformed_df['price']= transformed_df['price'].str.replace(',','').replace('Free',0).astype('float')
```
